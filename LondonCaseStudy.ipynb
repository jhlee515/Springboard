**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transpose DataFrame
t_prop = properties.T
display(t_prop.head())
print(t_prop.index)
```

```python
# Reset index to integers instead of borough names
t_prop = t_prop.reset_index()
t_prop.head()
```

```python
# Rename columns to first row of dataframe 
t_prop.columns = t_prop.loc[0]
t_prop = t_prop.drop(index = 0).reset_index(drop = True)
t_prop.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename certain columns
t_prop = t_prop.rename(columns = {'Unnamed: 0': 'boroughs', pd.NaT: 'id'})
t_prop.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt DataFrame to clean up
melted_prop = pd.melt(t_prop, id_vars = ['boroughs', 'id'])
print('Melted shape:' + str(melted_prop.shape))
display(melted_prop.head())
```

```python
# Rename melted_prop columns
melted_prop = melted_prop.rename(columns = {0: 'date', 'value': 'avg_price'})
melted_prop.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
melted_prop.dtypes # melted['avg_price'] returns 'object', must change to float
melted_prop['avg_price'] = pd.to_numeric(melted_prop['avg_price'])
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
print(melted_prop['boroughs'].unique()) # returns 48 unique boroughs when there should be only 32
print('---------------------------------------------------')
print(melted_prop[melted_prop['boroughs'] == 'SOUTH EAST'])
print('---------------------------------------------------')
print(melted_prop[melted_prop['boroughs'].str.contains('Unnamed: ')])

'''It seems that rows containing regions such as 'SOUTH EAST' actually contain information,
thus, these rows will not be dropped. However, rows containing 'Unnamed: ' only have null values
for 'id' and 'avg_price', therefore, these rows will be dropped.'''

melted_prop.dropna(subset = ['id', 'avg_price'], inplace = True)
```

After removing rows that contained 'Unnamed: ' and various regions such as 'NORTH EAST' and 'SOUTH WEST', I found that there were still 36 boroughs instead of 32. To investigate, I compared the edited list of boroughs with the boroughs listed on [Wikipedia](https://en.wikipedia.org/wiki/List_of_London_boroughs). I found that the 'City of London' is not actually a borough along with 'England', 'Inner London', and 'Outer London'. By doing so, I was able to extract the correct list of 32 London Boroughs.

```python
# Creating a new list of the 32 boroughs for analysis
exclude_regions = ['NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 'EAST MIDLANDS',
                  'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 'SOUTH WEST',
                  'England', 'Inner London', 'City of London', 'Outer London']

relevant_boroughs = list(set(melted_prop['boroughs'].unique()) - set(exclude_regions))
```

```python
print(melted_prop.shape)
print(melted_prop.head())
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Plot of a single borough (Bexley for example)
bexley = melted_prop[melted_prop['boroughs'] == 'Bexley']
plt.plot(bexley.date, bexley.avg_price)
plt.title('Average Housing Price in Bexley')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```

```python
# Plot of all boroughs/regions in dataset
for region in relevant_boroughs:
    rdf = melted_prop[melted_prop['boroughs'] == region]
    plt.plot(rdf.date, rdf.avg_price)
    
plt.title('Average Housing Price in various London boroughs')
plt.xlabel('Year')
plt.ylabel('Average price')
plt.tight_layout()
plt.savefig('AveragePrice.png', dpi = 300)
plt.show()
```

```python
# Creating a new DataFrame for yearly averaged housing prices by borough
averaged = melted_prop.groupby(['boroughs', 'year']).agg('mean')
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Function to calculate ratio of latest average housing price to oldest average housing price
def create_price_ratio(borough):
    prices = averaged.loc[borough, 'avg_price']
    ratio = prices.iloc[-1]/prices.iloc[0] #calculate ratio based on last and first recorded data points
    
    return ratio   
```

```python
# Create empty dictionary of boroughs for calculated ratios
ratios = dict.fromkeys(relevant_boroughs)
```

```python
# Calculate and fill in ratios for each borough
for borough in relevant_boroughs:
    ratios[borough] = create_price_ratio(borough)
    
ratios
```

```python
# Sort ratios dictionary by descending values
desc_ratios = {k: round(v,3) for k, v in sorted(ratios.items(), key=lambda x: x[1], reverse=True)}
desc_ratios
```

```python
boroughs = list(desc_ratios.keys())
values = list(desc_ratios.values())
plt.figure(figsize = (15,5))
plt.bar(boroughs, values)
plt.title('Ratio of Average Housing Prices from 1995 to 2020')
plt.xlabel('Boroughs')
plt.xticks(rotation = 90)
plt.tight_layout()
plt.savefig('RatioGraph.png', dpi = 300)
plt.show()
```

```python
overall_avg_increase = round(np.mean(values),3)
print('On average, housing prices in London have increase by a multiple of {}.'.format(str(overall_avg_increase)))
```
