**Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members!**

```python
q10 = '''
SELECT facility, SUM(revenue) as totalRevenue FROM
(SELECT f.name as facility, 
CASE 
    WHEN b.memid = 0 THEN b.slots*f.guestcost
    ELSE b.slots*f.membercost 
    END AS revenue
FROM Bookings as b
LEFT JOIN Facilities as f on b.facid = f.facid
LEFT JOIN Members as m on m.memid = b.memid
) AS sub
GROUP BY facility
HAVING totalRevenue < 1000
ORDER BY totalRevenue DESC
'''

execute(q10)
```

**Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order**

```python
q11 = '''
SELECT m.surname||', '|| m.firstname  AS name, Members.surname||', '|| Members.firstname  AS recommended
FROM Members AS m
INNER JOIN Members ON Members.memid = m.recommendedby
ORDER BY recommended
'''
execute(q11)
```

**Q12: Find the facilities with their usage by member, but not guests**

```python
q12 = '''
SELECT b.facid, f.name as facility, SUM(CASE WHEN memid != 0 THEN 1 ELSE 0 END) as total_usage
FROM Bookings as b
LEFT JOIN Facilities as f on b.facid = f.facid
GROUP BY b.facid
'''
execute(q12)
```

**Q13: Find the facilities usage by month, but not guests**

```python
q13 = '''
SELECT b.facid as facid, f.name as facility, strftime('%m', b.starttime) as month, SUM(CASE WHEN b.memid != 0 THEN 1 ELSE 0 END) as monthlyUsage
FROM Bookings as b
LEFT JOIN Facilities as f ON b.facid = f.facid
GROUP BY b.facid, month
'''
execute(q13)
```
